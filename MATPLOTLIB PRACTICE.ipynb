# 1. Subplotting

```python
a = np.linspace(0,10,100)
```

```python
plt.figure(figsize = (10,10))

plt.subplot(2,2,1)
plt.plot(a,np.sin(a))
plt.title("Sin")

plt.subplot(2,2,2)
plt.plot(a,np.cos(a))
plt.title("Cos")

plt.subplot(2,2,3)
plt.plot(a,np.tan(a))
plt.title("Tan")

```

stacked bar chart

```python
a = [1,2,3]
b = [4,5,6]
plt.plot(a,b,label = "l")
plt.title(label = "OUR FIRST GRAPH",fontdict = {"fontsize":20,"color":"violet"})
plt.xlabel("x axis",loc = "right")
plt.ylabel("y axis",loc = "top")
plt.xticks(np.arange(0,3), ['Tom', 'Dick', 'Sue'])
```

```python
a = [1,2,3]
b = [4,5,6]
plt.plot(a,b,label = "l",marker = ".",markersize = 10,markeredgecolor = "red")
plt.title(label = "OUR FIRST GRAPH",fontdict = {"fontsize":20,"color":"violet"})
plt.xlabel("x axis",loc = "right")
plt.ylabel("y axis",loc = "top")
```

```python
a = [1,2,3]
b = [4,5,6]
plt.plot(a,b,"b*-",label = "line")
plt.title(label = "OUR FIRST GRAPH",fontdict = {"fontsize":20,"color":"violet"})
plt.xlabel("x axis",loc = "right")
plt.ylabel("y axis",loc = "top")
plt.legend()

plt.savefig("first graph.png",dpi = 300)
```

# BOX PLOT
It is a graph which tells us all the outliers as well as the mean values of a particular data.

```python
data.head(3)
```

```python
plt.boxplot(data["tip"])
plt.ylabel("tips",loc = "top")
plt.xlabel("")
```

```python
plt.boxplot(data["total_bill"])
```

a boxplot is used to find outliers in a given data under a specific column . 
a boxplot consists of x axis, y axis , median(the line of a box), whiskers (the 
lines joining the box and the outliers)
